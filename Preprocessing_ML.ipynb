# Shapefile Data

```python
zones_gdf = gpd.read_file("taxi_zones/taxi_zones.shp")
zones_gdf.tail(3)
```

```python
Manh = zones_gdf[zones_gdf['borough'] == 'Manhattan']
Manh_list = Manh['LocationID'].tolist()
print(Manh.shape)
Manh.head(3)
```

# Convert to Dataframe

```python
Full_df = pd.concat([Aug_df, Sep_df, Oct_df, Nov_df, Dec_df, Jan_df], axis = 0)
print(Full_df.shape)
Full_df.head(3)
```

```python
Manh_all = Full_df[Full_df['PULocationID'].isin(Manh_list)]
Manh_all = Manh_all[['tpep_pickup_datetime', 'PULocationID']]
```

```python
Manh_all['pickup_hour'] = Manh_all['tpep_pickup_datetime'].dt.floor('H')
Manh_all.tail(3)
```

```python
Manh_all['year'] = Manh_all['tpep_pickup_datetime'].dt.year
Manh_all.tail(3)
```

```python
#Make sure we only have data in the past few years
print(Manh_all['year'].unique())
Manh_all = Manh_all[Manh_all['year'].isin([2025, 2024])]
print(Manh_all['year'].unique())
```

```python
grouped_df = Manh_all.groupby(['PULocationID', 'pickup_hour']).size().reset_index(name='Count')
print(grouped_df.shape)
grouped_df.head(10)
```

## Making sure Counts are zero for hours with no pick-ups

```python
# All daytime list
all_hours = pd.date_range(start='2024-08-01 00:00:00',
                          end='2025-01-31 23:00:00',
                          freq='H')


grouped_df = grouped_df[grouped_df['pickup_hour'].isin(all_hours)]
```

```python
all_index = pd.MultiIndex.from_product(
    [Manh_list, all_hours],
    names=['PULocationID', 'pickup_hour']
)
print(len(all_index))
all_index[:10]
```

```python
grouped_df = grouped_df.set_index(['PULocationID', 'pickup_hour',])
grouped_df.head(10)
```

```python
grouped_df = grouped_df.reindex(all_index, fill_value=0).reset_index()
grouped_df.head(10)
```

```python
#Make a separate column of US holiday
us_holidays = holidays.UnitedStates(years=[2024, 2025])
grouped_df['is_holiday'] = grouped_df['pickup_hour'].dt.date.astype('datetime64[ns]').isin(us_holidays)
grouped_df['is_holiday'] =  grouped_df['is_holiday'].astype(int)
grouped_df.tail(3)
```

```python
# Step 3: Adding time features
grouped_df['hour'] = grouped_df['pickup_hour'].dt.hour
grouped_df['dayofweek'] = grouped_df['pickup_hour'].dt.dayofweek  # 0=Mon, 6=Sun
grouped_df['is_weekend'] = grouped_df['dayofweek'].isin([5,6]).astype(int)
grouped_df.sort_values(['PULocationID', 'pickup_hour'], inplace=True)
grouped_df.head(5)
```

# Creating lagged hours

```python
#Creating lagged 1h and lagged 24 hour, lagged 1 week
grouped_df['lag_1h'] = grouped_df.groupby('PULocationID')['Count'].shift(1)
grouped_df['lag_24h'] = grouped_df.groupby('PULocationID')['Count'].shift(24)
grouped_df['lag_1w'] = grouped_df.groupby('PULocationID')['Count'].shift(168)
grouped_df[['lag_1h', 'lag_24h', 'lag_1w']] = grouped_df[['lag_1h', 'lag_24h', 'lag_1w']].fillna(0)

#3 hour rolling average
grouped_df['rolling_avg_3h'] = grouped_df.groupby('PULocationID')['Count'].rolling(3).mean().shift(1).reset_index(0, drop=True)
grouped_df.tail(5)
```

```python
dummies = pd.get_dummies(grouped_df['PULocationID'], prefix='zone', drop_first=True).astype(int)
Processed = pd.concat([grouped_df, dummies], axis=1)
Processed.tail(3)
```
